```python
#!/usr/bin/python3.9 
# -*- coding: utf-8 -*-  
#author Kunlun Zhu 2022/6/23
#setup the python environment
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Let's suppose a 10 nodes network
A = np.random.randint(2, size=(10,10)) #Adjacency matrix
link_list = []
for i in range(10):
    A[i][i] = 0
for index, a in np.ndenumerate(A):
    if a == 1:
        link_list.append((index[0], index[1]))
```

```python
G = nx.DiGraph()
G.add_nodes_from(range(10))
for item in link_list:
    G.add_edge(item[0], item[1], weight = np.random.rand())
nx.draw(G, with_labels=True)
print('Citation Network')
plt.show()
```
